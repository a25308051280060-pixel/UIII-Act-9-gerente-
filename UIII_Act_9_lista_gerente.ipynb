```python
# --- ASISTENTE IA PARA GERENCIA DE PROYECTOS v1.0 ---

# Inicialización de listas vacías al inicio [5]
nombres_proyectos = []
presupuestos = []
prioridades = []

ejecutar_menu = "si" # Variable de control para el bucle principal [6]

print("--- BIENVENIDO, GERENTE. INICIANDO SISTEMA DE IA ---")

# Bucle principal interactivo utilizando 'while' [2, 7]
while ejecutar_menu == "si":
    print("\n" + "="*40)
    print("MENÚ DE GESTIÓN GERENCIAL:")
    print("1. Registrar nuevo proyecto")
    print("2. Mostrar todos los proyectos")
    print("3. Ejecutar Análisis de IA (Simulación)")
    print("4. Salir del sistema")
    print("="*40)

    opcion = input("Seleccione una opción (1-4): ")

    # ESTRUCTURA DE CONTROL SELECTIVA (if-elif-else) [1, 8]
    if opcion == "1":
        # --- REGISTRO DE DATOS ---
        print("\n[MODO REGISTRO]")
        nombre = input("Nombre del proyecto: ")
        dinero = float(input("Presupuesto asignado ($): "))
        nivel = int(input("Nivel de importancia (1-10): "))

        # Agregar elementos a las listas usando append() [5]
        nombres_proyectos.append(nombre)
        presupuestos.append(dinero)
        prioridades.append(nivel)
        print("¡Proyecto guardado en la base de datos!")

    elif opcion == "2":
        # --- LISTADO MANUAL CON WHILE ---
        print("\n[LISTA DE PROYECTOS ACTUALES]")

        # Verificamos si la lista está vacía [9, 10]
        if len(nombres_proyectos) == 0:
            print("No hay proyectos registrados aún.")
        else:
            indice = 0 # Iniciamos contador para recorrer la lista [7, 11]
            while indice < len(nombres_proyectos):
                print(f"{indice + 1}. {nombres_proyectos[indice]} | Presupuesto: ${presupuestos[indice]}")
                indice = indice + 1 # Incremento manual (no usamos for) [12]

    elif opcion == "3":
        # --- SIMULACIÓN DE INTELIGENCIA ARTIFICIAL ---
        print("\n[PROCESANDO ANÁLISIS INTELIGENTE DE IA...]")

        if len(nombres_proyectos) == 0:
            print("Error IA: No hay datos suficientes para analizar.")
        else:
            i = 0
            while i < len(nombres_proyectos):
                # Extraemos datos de las listas para la lógica de la IA [5]
                p_nombre = nombres_proyectos[i]
                p_presupuesto = presupuestos[i]
                p_prioridad = prioridades[i]

                # LÓGICA IA: Uso de operadores lógicos y relacionales [13, 14]
                # CASO 1: Alta prioridad y presupuesto robusto (Inversión VIP)
                if p_prioridad >= 8 and p_presupuesto > 5000:
                    decision = "DESPLIEGUE INMEDIATO (Prioridad Máxima)"

                # CASO 2: Presupuesto bajo pero prioridad media (Optimización)
                elif p_presupuesto <= 1000 or p_prioridad < 5:
                    decision = "REVISIÓN DE COSTOS (Baja Viabilidad)"

                # CASO 3: Proyectos equilibrados
                else:
                    decision = "EN ESPERA DE ASIGNACIÓN (Estable)"

                print(f"Proyecto: {p_nombre} --> Diagnóstico IA: {decision}")
                i = i + 1

    elif opcion == "4":
        # Opción para terminar el programa [6]
        ejecutar_menu = "no"
        print("IA: Desconectando. ¡Buen día, Gerente!")

    else:
        print("Opción no válida. Intente de nuevo.")

print("\n--- PROGRAMA FINALIZADO ---")
print("programa terminado por Ambar Fierro")
```

```python
# --- SISTEMA DE IA PARA EVALUACIÓN DE GERENTES v1.0 ---

# Al inicio, la lista de gerentes está vacía [1, 2]
lista_gerentes = []

print("--- BIENVENIDO AL CENTRO DE MANDO DE IA GERENCIAL ---")

# Bucle infinito para el menú interactivo [3]
while True:
    print("\n" + "="*40)
    print("MENÚ DE GESTIÓN:")
    print("1. Registrar nuevo Gerente")
    print("2. Listar Gerentes registrados")
    print("3. Ejecutar Análisis de IA (Desempeño)")
    print("4. Salir del sistema")
    print("="*40)

    opcion = input("Seleccione una opción (1-4): ")

    # Estructura selectiva para el menú [4, 5]
    if opcion == "1":
        # Al inicio de cada registro, el diccionario está vacío [6]
        gerente_datos = {}

        print("\n[MODO REGISTRO]")
        # Interatividad: Captura de datos [7]
        nombre = input("Nombre del Gerente: ")
        ventas = float(input("Ventas mensuales logradas ($): "))
        eficiencia = int(input("Nivel de eficiencia operativa (1-100): "))

        # Guardar datos en el diccionario utilizando pares clave-valor [8]
        gerente_datos["nombre"] = nombre
        gerente_datos["ventas"] = ventas
        gerente_datos["eficiencia"] = eficiencia

        # Agregar el diccionario a la lista [1]
        lista_gerentes.append(gerente_datos)
        print(f"¡Gerente {nombre} registrado exitosamente!")

    elif opcion == "2":
        print("\n[LISTADO DE GERENTES]")
        # Verificación de lista vacía [9]
        if not lista_gerentes:
            print("No hay gerentes en la base de datos.")
        else:
            # Ciclo for para recorrer la lista de diccionarios [10, 11]
            for g in lista_gerentes:
                print(f"Gerente: {g['nombre']} | Ventas: ${g['ventas']} | Eficiencia: {g['eficiencia']}%")

    elif opcion == "3":
        print("\n[PROCESANDO DIAGNÓSTICO DE IA...]")
        if not lista_gerentes:
            print("IA Alerta: No hay datos suficientes para analizar.")
        else:
            # La IA recorre la lista y evalúa cada diccionario [10, 12]
            for g in lista_gerentes:
                # Lógica de IA: Uso de operadores lógicos y relacionales [13, 14]

                # Caso 1: Alto desempeño (Ventas > 5000 Y Eficiencia >= 90)
                if g["ventas"] > 5000 and g["eficiencia"] >= 90:
                    diagnostico = "LIDERAZGO EXCEPCIONAL - Candidato a CEO"

                # Caso 2: Desempeño promedio (Ventas entre 2000 y 5000 O Eficiencia > 70)
                elif (2000 <= g["ventas"] <= 5000) or g["eficiencia"] > 70:
                    diagnostico = "ESTABLE - Mantener en capacitación"

                # Caso 3: Desempeño bajo
                else:
                    diagnostico = "CRÍTICO - Plan de mejora necesario"

                print(f"Resultado IA para {g['nombre']}: {diagnostico}")

    elif opcion == "4":
        print("Cerrando sistema de IA... ¡Hasta pronto!")
        break # Rompe el ciclo while True para salir [15, 16]

    else:
        print("Error: Opción no válida. Intente de nuevo.")

print("\n--- PROGRAMA FINALIZADO ---")
print("programa terminado por Ambar Fierro")
```
